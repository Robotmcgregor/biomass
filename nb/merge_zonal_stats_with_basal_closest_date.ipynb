# Merged zonal stats with basal closest date

This notebook collates the outputs from “Merged basal zonal stats collation” renames columns with feature identifies and returns only the closest features to the field data and exported as a csv to the output directory.

The following conditions apply:

 - run after merged_basal_zonal_stats_colation notebook.
 - env = base

```python
import pandas as pd
from glob import glob
import os
from calendar import monthrange
from datetime import datetime
```

```python
dir_ = r"F:\cdu\data\output\zonal_stats\merged_basal_zonal"
output_dir = r"F:\cdu\data\output\zonal_stats\merged_all_data"
```

```python
df_list = []

for file_ in glob(os.path.join(dir_, "*.csv")):
    df = pd.read_csv(file_)
    df_list.append(df)
```

```python
df = df_list[0]
df.columns
```

```python
def rename_cols(df):
    
#     df1 = df_list[0]
    name = df["im_name"][0]
    name_split = name.split("_")
    file = name_split[-1]
    print(file)
    file_split = file.split(".")
    final_file = file_split[0]
    print(final_file)

    dict_ = {'im_name': 'im_name'.replace("im", final_file),
             'im_s_date': 'im_s_date'.replace("im", final_file),
             'im_e_date': 'im_e_date'.replace("im", final_file),
             'im_s_d': 'im_s_d'.replace("im", final_file),
             'im_e_d': 'im_e_d'.replace("im", final_file)}

    print(dict_)
    
    # call rename () method
    df.rename(columns=dict_,
              inplace=True)

    print(df.columns)
    
    return df


```

```python
df1 = rename_cols(df_list[0])

for df in df_list[1:]:
    print(df.columns)
    df = rename_cols(df)
    print(df)
    df1= pd.merge(df1, df, on=['uid', 'site', 'date', 'lon_gda94', 'lat_gda94', 'bio_l_kg1ha',
       'bio_t_kg1ha', 'bio_b_kg1ha', 'bio_w_kg1ha', 'bio_br_kg1ha',
       'bio_s_kg1ha', 'bio_r_kg1ha', 'bio_agb_kg1ha', 'c_l_kg1ha', 'c_t_kg1ha',
       'c_b_kg1ha', 'c_w_kg1ha', 'c_br_kg1ha', 'c_s_kg1ha', 'c_r_kg1ha',
       'c_agb_kg1ha', 'date_d', 'ident'], how= 'left')
    print(df1)
    
```

```python
df1.to_csv(os.path.join(output_dir, "merged_all_data.csv"), index=False)
```
