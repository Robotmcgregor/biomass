# Select only variables for model

This notebook selects out the variables from the output created from “Merged zonal stats with basal closest date” select the variable outputs listed in the keep list and exports as a csv to the output directory.

The following conditions apply:

 - run after merged_zonal stats with basal closest date notebook.
 - env = base

```python
import pandas as pd
from glob import glob
import os
```

```python
dir_ = r"F:\cdu\data\output\zonal_stats\merged_all_data"
output_dir = r"F:\cdu\data\output\zonal_stats\variables_only"
```

```python
for file_ in glob(os.path.join(dir_, "*.csv")):
    df = pd.read_csv(file_)
```

```python
column_list = df.columns.tolist()
```

```python
# for i in column_list:
#     print(i)
```

```python
want_list = ["uid", "site", "lon_gda94", "lat_gda94", "bio_agb_kg1ha", "mean", "p25", "p50", "p75", "min", "max"]
```

```python
header = [ele for ele in column_list for x in want_list if x in ele]
```

```python
date_list = df['date'].tolist()
```

```python
df1 = df[header]
df1.insert(2, "date", date_list)
```

```python
df1.to_csv(os.path.join(output_dir, "variables_only_agb.csv"), index=False)
```
