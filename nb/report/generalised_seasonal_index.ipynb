# Calculate seasonal indecies

```python
import numpy as np
import pandas as pd

# Sample data: Monthly rainfall data for 3 years
data = {
    'Year': [2018, 2018, 2018, 2018, 2018, 2018, 2018, 2018, 2018, 2018, 2018, 2018,
             2019, 2019, 2019, 2019, 2019, 2019, 2019, 2019, 2019, 2019, 2019, 2019,
             2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020],
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']*3,
    'Rainfall': [78, 64, 59, 47, 39, 42, 35, 30, 45, 55, 60, 70,
                 85, 70, 65, 50, 40, 45, 38, 33, 48, 57, 62, 75,
                 80, 66, 62, 49, 41, 44, 36, 31, 46, 56, 61, 72]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate the monthly average rainfall
monthly_avg = df.groupby('Month')['Rainfall'].mean()

# Calculate the overall monthly average rainfall
overall_monthly_avg = df['Rainfall'].mean()

# Calculate the seasonal indices
seasonal_indices = (monthly_avg / overall_monthly_avg) * 100

# Display the results
seasonal_indices = seasonal_indices.reset_index().rename(columns={'Rainfall': 'Seasonal Index'})
print(seasonal_indices)
```
