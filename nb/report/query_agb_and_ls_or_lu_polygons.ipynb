```python
points = gpd.read_file(r"X:\PGB\RSU\biomass\collate_agb\20240318\biomass.shp", driver="ESRI shapefile")
```

```python
points_gdf = gpd.read_file(r"X:\PGB\RSU\biomass\collate_agb\20240318\biomass.shp", driver="ESRI shapefile")

list_a = []

for (root,dirs,files) in os.walk(path_, topdown=True):
#     print (root)
#     print (dirs)
#     print (files)
    for f in files:
        if f.endswith(".shp"):
            print(f)
            shp = os.path.join(root, f)
            print(shp)
            print("-"*20)
            try:
                polygons_gdf = gpd.read_file(shp, driver="ESRI shapefile")
                print(polygons_gdf.crs)

                # Perform spatial join
                joined_gdf = gpd.sjoin(points_gdf, polygons_gdf, how='left', op='within')
                joined_gdf.dropna(inplace = True)

                if not joined_gdf.empty:

                    p, s = os.path.split(shp)
                    sp = s.replace(".shp", "_agb.shp")
                    out = os.path.join(outdir, sp)
                    joined_gdf.to_file(out)
                    
                    df = pd.DataFrame(joined_gdf)
                    list_a.append(df)
            except:
                print("pass")
            
df = pd.concat(list_a)
```
