#### Import modules

```python
import pandas as pd
import os
import geopandas as gpd
```

```python
drive = "F"
date_slats = "20230326"
date_tern = "20230326"
date_bio = "20230326"
date_blue = "20230326"

date_output = '20230326'
```

#### Inputs

```python
# Slats data
slats_path = r"{0}:\cdu\data\output\{1}\slats_c_bio_site_totals.csv".format(drive, date_slats)
# TERN data
tern_path = r"{0}:\cdu\data\output\{1}\c_tern_bio_site_totals.csv".format(drive, date_tern)
# Tern Portal biomass - biolib
bio_path = r"{0}:\cdu\data\output\{1}\biolib_c_bio_site_totals.csv".format(drive, date_bio)
# Blue Folder biomass
blue_path = r"{0}:\cdu\data\output\{1}\blue_folders_c_bio_site_totals.csv".format(drive, date_blue)
```

```python
output_dir = r"{0}:\cdu\data\output\{1}".format(drive, date_output)
u_output_dir = r"X:\PGB\RSU\biomass\collated_agb\{1}".format(drive, date_output)
```

#### Outputs

```python
slats = pd.read_csv(slats_path)
tern = pd.read_csv(tern_path)
bio = pd.read_csv(bio_path)
blue = pd.read_csv(blue_path)
```

```python
print(f"slats shape {slats.shape}")
print(f"tern shape {tern.shape}")
print(f"biolib shape {bio.shape}")
print(f"blue shape {blue.shape}")
```

```python
df = pd.concat([slats, tern, bio, blue])
```

```python
df.shape
```

```python
df
```

```python
site_new = []
for index, row in df.iterrows():
    site = row.site
    date = str(row.date)

    site_new.append(f"{site.lower()}_{str(date[:4])}")

df.rename(columns={"site": "site_name"}, inplace=True)
df.insert(1, "site", site_new)
```

```python
df.sort_values(by=['site'], inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
df["uid"] = df.index + 1
df
```

```python
#Count duplicate on certain columns
df.duplicated(subset=['site']).sum()
```

```python
# Extract duplicate rows
df.loc[df.duplicated(subset=["site"]), :]
```

```python
df.sample(20)
```

```python
out = os.path.join(output_dir, "slats_tern_blue_biolib_biomass.csv")

df.to_csv(os.path.join(out), index = False)
print("file exported: ", out)
```

```python
out = os.path.join(u_output_dir, "slats_tern_blue_biolib_biomass.csv")

df.to_csv(os.path.join(out), index = False)
print("file exported: ", out)
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.lon_gda94, df.lat_gda94))
```

```python
gdf.set_crs("epsg:4283", inplace=True)
```

```python
print(gdf.crs)
```

```python
gdf.to_file(os.path.join(output_dir, "slats_tern_blue_biolib_biomass.shp"), driver = "ESRI Shapefile")
gdf.to_file(os.path.join(u_output_dir, "slats_tern_blue_biolib_biomass.shp"), driver = "ESRI Shapefile")
```
