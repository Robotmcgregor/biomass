#### Outputs

```python
slats = pd.read_csv(slats_path)
tern = pd.read_csv(tern_path)
```

```python
slats.shape
```

```python
sorted(slats.site.unique())
```

```python
tern.shape
```

```python
df = pd.concat([slats, tern])
```

```python
df.shape
```

```python
# Convert the value of 'Name' column to lowercase
df['site'] = df['site'].str.lower()
```

```python
df.sort_values("site", inplace = True)
```

```python
df.reset_index(inplace=True, drop=True)
df["uid"] = df.index + 1
df
```

```python
out = os.path.join(output_dir, "slats_tern_biomass.csv")

df.to_csv(os.path.join(out), index = False)
print("file exported: ", out)
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.lon_gda94, df.lat_gda94))
```

```python
gdf.set_crs("epsg:4283", inplace=True)
```

```python
print(gdf.crs)
```

```python
gdf.to_file(os.path.join(output_dir, "slats_tern_biomass.shp"), driver = "ESRI Shapefile")
```
