#### Outputs

```python
slats = pd.read_csv(slats_path)
tern = pd.read_csv(tern_path)
#bio = pd.read_csv(bio_path)
#blue = pd.read_csv(blue_path)
```

```python
print(f"slats shape {slats.shape}")
print(f"tern shape {tern.shape}")
#print(f"biolib shape {bio.shape}")
#print(f"blue shape {blue.shape}")
```

```python
tern.shape
```

```python
df = pd.concat([slats, tern])#, bio, blue])
```

```python
df.shape
```

```python
df
```

```python
site_new = []
siteyear_list = []
for index, row in df.iterrows():
    site = row.site
    date = str(row.date)

    site_new.append(f"{site.lower()}_{str(date[:4])}")

df.rename(columns={"site": "site_name"}, inplace=True)
df.insert(1, "site", site_new)
```

```python
df
```

```python
df.sort_values(by=['site'], inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
df["uid"] = df.index + 1
df
```

```python
#Count duplicate on certain columns
df.duplicated(subset=['site']).sum()
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.lon_gda94, df.lat_gda94))
```
