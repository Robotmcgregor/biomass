```python
import pandas as pd
```

```python
dbi = r"C:\Users\robot\projects\biomass\collated_zonal_stats\dry_mask\dbi_mask_dry.csv"
dp1 = r"C:\Users\robot\projects\biomass\collated_zonal_stats\dry_mask\dp1_mask_dry.csv"
dbi_dp1 = r"C:\Users\robot\projects\biomass\collated_zonal_stats\dry_mask\dbi_dp1_mask_dry.csv"
```

```python
dbi_df = pd.read_csv(dbi)
dp1_df = pd.read_csv(dp1)
dbi_dp1_df = pd.read_csv(dbi_dp1)
```

```python
def list_site(df, c_name):
    df.sort_values(by="uid", inplace=True)
    l = df[c_name].tolist()
    return l    

def compare_lists(list1, list2, list3):
    differences = []
    max_length = max(len(list1), len(list2), len(list3))
    
    for i in range(max_length):
        elem1 = list1[i] if i < len(list1) else None
        elem2 = list2[i] if i < len(list2) else None
        elem3 = list3[i] if i < len(list3) else None
        
        if elem1 != elem2 or elem1 != elem3 or elem2 != elem3:
            differences.append((i, elem1, elem2, elem3))
    
    return differences
```

```python
dbi_list = list_site(dbi_df, "site_clean")
dp1_list = list_site(dp1_df, "site_clean")
dbi_dp1_list = list_site(dbi_dp1_df, "site_clean")
```

```python
differences = compare_lists(dbi_list, dp1_list, dbi_dp1_list)
```

```python
if differences:
    print("The lists have differences at the following positions (index, list1, list2, list3):")
    for diff in differences:
        print(diff)
else:
    print("All three lists have the same elements in the same order.")
```

```python
def compare_sets(list1, list2, list3):
    set1 = set(list1)
    set2 = set(list2)
    set3 = set(list3)
    
    unique_to_list1 = set1 - set2 - set3
    unique_to_list2 = set2 - set1 - set3
    unique_to_list3 = set3 - set1 - set2
    
    differences = {
        'unique_to_list1': unique_to_list1,
        'unique_to_list2': unique_to_list2,
        'unique_to_list3': unique_to_list3
    }
    
    return differences


differences = compare_sets(dbi_list, dp1_list, dbi_dp1_list)

print("Elements unique to list1:", differences['unique_to_list1'])
print("Elements unique to list2:", differences['unique_to_list2'])
print("Elements unique to list3:", differences['unique_to_list3'])
```
