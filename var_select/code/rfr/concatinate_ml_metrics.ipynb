```python
import os
import pandas as pd
import math

# Directory path to search through
directory_path = r'H:\biomass\model_all01_02_03_04_05_06_v2'

# List to store the DataFrames
df_list = []

# Walk through the directory and subdirectories
for root, dirs, files in os.walk(directory_path):
    for d in dirs:
        full_dir_path = os.path.join(root, d)
        print("Checking directory:", full_dir_path)  # Debug: Print the full path being checked
        
        if d.lower().startswith(("all")):  # Use .lower() to ignore case sensitivity
            dir_path = os.path.join(root, d)  # Get the full path of the directory
            print("Matched directory:", dir_path)  # Debug: Directory matched
            
            # Walk through the directory and subdirectories
            for sub_root, sub_dirs, sub_files in os.walk(dir_path):
                for f in sub_files:
                    full_file_path = os.path.join(sub_root, f)
                    print("Checking file:", full_file_path)  # Debug: Print the full path being checked
        
                    if f.endswith(("retest_metrics.csv", "validate_metrics.csv")):
                        print(" - Found file:", full_file_path)  # Debug: Print the file path
                        df = pd.read_csv(full_file_path)

                        # Convert r2 and rmse columns from scientific notation to float
                        df['r2'] = df['r2'].astype(float)
                        df['rmse'] = df['rmse'].astype(float)

                        df_list.append(df)

# Now df_list contains all the DataFrames from matched files
```

```python
import pandas as pd# Concatenate all DataFrames in the list
combined_df = pd.concat(df_list, ignore_index=True)
#combined_df['file'] = combined_df.apply(lambda row: f"{row['var']}_{row['fac']}_{row['mdl']}_sel_{int(row['sel_num'])}_variable_score.csv", axis=1)

combined_df['file'] = combined_df.apply(
    lambda row: f"{row['var']}_{row['fac']}_{row['mdl']}_sel_{int(row['sel_num']) if not math.isnan(row['sel_num']) else ''}_variable_score.csv",
    axis=1
)
```

```python
def create_sorted_dfs_and_export(df, output_directory):
    """
    Create sorted DataFrames for each combination of 'mdl' and a new feature created from the first 3 characters of 'var',
    showing the top 20 rows per feature, and export them as CSV files.

    Parameters:
    - df: DataFrame, the input DataFrame that is already sorted.
    - output_directory: str, path to the directory where the CSV files will be saved.
    """
    # Ensure the output directory exists
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Create a new feature from the first 3 characters of 'var'
    df['var_prefix'] = df['var'].str[:3]

    # Group by 'mdl' and the new feature 'var_prefix'
    grouped = df.groupby(['mdl', 'var_prefix'])

    # Process each group
    for (mdl, var_prefix), group_df in grouped:
        # Sort each group by the desired columns (assuming df is pre-sorted)
        sorted_group = group_df.head(20)  # Take the top 20 rows

        # Create a filename based on the group keys
        filename = f"{mdl}_{var_prefix}_top20.csv"
        sub_path = os.path.join(output_directory, f"{mdl}_{var_prefix}")
        if not os.path.isdir(sub_path):
            os.mkdir(sub_path)
        file_path = os.path.join(sub_path, filename)

        # Export the sorted DataFrame to CSV
        sorted_group.to_csv(file_path, index=False)

        print(f"Exported: {file_path}")

# Example usage
# Assuming df is your sorted DataFrame
output_directory = r"H:\\biomass\\model_analysis\\"
create_sorted_dfs_and_export(sort_index, output_directory)
```

```python
def create_sorted_dfs_and_export(df, output_directory):
    """
    Create sorted DataFrames for each combination of 'mdl', 'var', 'data', and 'stats', 
    showing the top 20 rows per feature, and export them as CSV files.

    Parameters:
    - df: DataFrame, the input DataFrame that is already sorted.
    - output_directory: str, path to the directory where the CSV files will be saved.
    """
    # Ensure the output directory exists
    import os
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Group by 'mdl', 'var', 'data', and 'stats'
    grouped = df.groupby(['mdl', 'var', 'data', 'stats'])

    # Process each group
    for (mdl, var, data, stats), group_df in grouped:
        # Sort each group by the desired columns (assuming df is pre-sorted)
        sorted_group = group_df.head(20)  # Take the top 20 rows

        # Create a filename based on the group keys
        filename = f"{mdl}_{var}_{data}_{stats}_top20.csv"
        sub_path = os.path.join(output_directory, mdl)
        if not os.path.isdir(sub_path):
            os.mkdir(sub_path)
        file_path = os.path.join(sub_path, filename)

        # Export the sorted DataFrame to CSV
        sorted_group.to_csv(file_path, index=False)

        print(f"Exported: {file_path}")

# Example usage
# Assuming df is your sorted DataFrame
output_directory = r"H:\\biomass\\model_analysis\\"
create_sorted_dfs_and_export(sort_index, output_directory)
```

```python
from bokeh.plotting import figure, output_file, save
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10

# Assuming df is your DataFrame
df['mdl_var_data_stats'] = df['mdl'] + '_' + df['var'] + '_' + df['data'] + '_' + df['stats']

# Create a list of unique mdl values
mdl_list = df['mdl'].unique().tolist()

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(data=dict(
    x=df['rmse_val'],
    y=df['r2_val'],
    mdl_var_data_stats=df['mdl_var_data_stats'],
    mdl=df['mdl'],
    var=df['var'],
    data=df['data'],
    stats=df['stats']
))

# Initialize the Bokeh plot
p = figure(plot_width=800, plot_height=600, title="Validation RMSE vs R2 Across Models and Variable Prefixes",
           tools="pan,wheel_zoom,box_zoom,reset")

# Create the scatter plot with color mapping based on mdl
p.circle('x', 'y', size=10, source=source, color=factor_cmap('mdl', palette=Category10[len(mdl_list)], factors=mdl_list), alpha=0.6)

# Add Hover tool
hover = HoverTool()
hover.tooltips = [
    ("(RMSE_X, R2_X)", "(@x, @y)"),
    ("Model_Var_Data_Stats", "@mdl_var_data_stats"),
    ("Model", "@mdl"),
    ("Variable", "@var"),
    ("Data", "@data"),
    ("Stats", "@stats"),
]
p.add_tools(hover)

# Set axis labels
p.xaxis.axis_label = 'RMSE_X'
p.yaxis.axis_label = 'R2_X'
# Save the plot to an HTML file
output_location = os.path.join(output_directory, "rmse_x_vs_r2_x_plot.html")
output_file(output_location)
save(p)

print("Plot saved as: ", output_location)
```
